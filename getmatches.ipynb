```python
def get_h2h(home_id,away_id):
    END_POINT = "https://api-football-v1.p.rapidapi.com/v3/fixtures/headtohead"
    
    querystring = {"h2h":"{}-{}".format(home_id, away_id)}

    
    headers = {
    	"X-RapidAPI-Key": "45e5b21a11mshd54d7307c0ebd2ap149128jsn23e494fd38af",
    	"X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
    }
    
    response = requests.get(END_POINT, headers=headers, params=querystring)
    
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        # Access the head-to-head statistics
        h2h_data = data['response']
        
        # Print the head-to-head statistics
        print("Head-to-Head Statistics:")
        print(h2h_data)
    else:
        # Print an error message if the request failed
        print("Error:", response.status_code)
    return h2h_data
```
