```python
import requests
```

```python
#get all league id

url = "https://api-football-v1.p.rapidapi.com/v3/leagues"

headers = {
	"X-RapidAPI-Key": "45e5b21a11mshd54d7307c0ebd2ap149128jsn23e494fd38af",
	"X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

# print(response.json())
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    # Access the list of leagues
    leagues = data['response']
    
    league_data = []
    for league in leagues:
        league_id = league['league']['id']
        league_name = league['league']['name']
        league_type = league['league']['type']
        league_logo = league['league']['logo']
        league_data.append({"ID": league_id, "Name": league_name, "Type": league_type, "Logo": league_logo})
        print(f"ID: {league_id}, Name: {league_name}, Type: {league_type}, Logo: {league_logo}")
    else:
        # Print an error message if the request failed
        print("Error:", response.status_code)
```
